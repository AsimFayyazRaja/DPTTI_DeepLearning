```python
# get and preprocess the data for task1

import pandas as pd
import numpy as np
import json
import itertools

path="SEMEVAL-2021-task6-corpus/data/"
```

```python
import json

p1=path+"training_set_task2.txt"

with open(p1) as input:
    train = json.loads(input.read())
```

```python
train
```

```python
texts=[]
techniques=[]
text_fragments=[]
start_ends=[]
ids=[]
all_labels=[]
all_labels.append('')

for val in train:
  try:
    texts.append(val['text'])
    ids.append(val['id'])
    sub_tech=[]
    sub_start=[]
    sub_text=[]
    for lab in val['labels']:
      all_labels.append(lab['technique'])
      sub_tech.append(lab['technique'])
      sub_text.append(lab['text_fragment'])
      sub_start.append([lab['start'],lab['end']])
    start_ends.append(sub_start)
    text_fragments.append(sub_text)
    techniques.append(sub_tech)
    
  except:
    print("val: ", val)
    print("-----------------")


all_labels=list(set(all_labels))
print(texts[:3])
print(techniques[:3])
print(text_fragments[:3])
print(start_ends[:3])
print(ids[:3])
```

```python
all_labels
```

```python
import copy
one_hot=np.zeros((len(all_labels),), dtype=int).tolist()
labs_dict={}
for i,label in enumerate(all_labels):
  lab=copy.deepcopy(one_hot)
  lab[i]=1
  labs_dict[label]=np.array(lab)
labs_dict
```

```python
len(all_labels)
```

```python
print(len(texts))
print(len(techniques))
print(len(text_fragments))
print(len(start_ends))
```

```python
features=[]
labels1=[]
labels2=[]
for i,text in enumerate(texts):
  #text=text.replace('\n',' ')
  #text=" ".join(text.split())
  if techniques[i]!=[]:
    for j,tech in enumerate(techniques[i]):
      lab1=labs_dict[tech]
      lab2=start_ends[i][j]
      features.append(text)
      labels1.append(lab1)
      labels2.append(lab2)
  else:
    features.append(text)
    labels1.append(labs_dict[''])
    labels2.append([0,0])


print(features[:5])
print(labels1[:5])
print(labels2[:5])
```

```python
import pickle

labels1=np.array(labels1)
labels2=np.array(labels2)

print(labels1.shape)
print(labels2.shape)

with open('original_training_task2_features.pkl', 'wb') as handle:
    pickle.dump(features, handle)

with open('original_training_task2_labels1.pkl', 'wb') as handle:
    pickle.dump(labels1, handle)

with open('original_training_task2_labels2.pkl', 'wb') as handle:
    pickle.dump(labels2, handle)

with open('task2_all_labels.pkl', 'wb') as handle:
    pickle.dump(all_labels, handle)

with open('task2_labels_dict.pkl', 'wb') as handle:
    pickle.dump(labs_dict, handle)
```
