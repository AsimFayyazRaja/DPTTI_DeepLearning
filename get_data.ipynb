```python
# get and preprocess the data for task1

import pandas as pd
import numpy as np
import json
import itertools

path="SEMEVAL-2021-task6-corpus/data/"
```

```python
def read_text_file(path):
  "returns processed list from the given json of text file"
  train_file=open(path)
  x=train_file.readline()
  i=0
  arr=[]
  data={}
  while(x):
    x=x.strip()
    if "id" in x:
      data['id']=x.split(':')[1]
    elif "labels" in x:
        x=train_file.readline()
        j=0
        while(x):
          x=x.strip()
          if "text" in x or '[]' in x:
            if j==0: # there is no label given
              data['labels']=""
            break
          try:
            data['labels']+=x
          except:
            data['labels']=x
          x=train_file.readline()
          j+=1
    if "text" in x:
      try:
        data['text']=x.split('"text":')[1]
      except:
        data['text']=x
      arr.append(data) 
      data={}
    x=train_file.readline()
  
```

```python
def clean_data(arr):
  "cleans the data"
  all_labels=[]
  for obj in arr:
    if obj['labels']!='':
      obj['labels']=obj['labels'].replace('],','')
      obj['labels']=obj['labels'].replace('"','')

      
      labs=obj['labels'].split(',')
      obj['labels']=labs
      all_labels.append(labs)
    else:
      labs=obj['labels'].split(',')
      obj['labels']=labs
      all_labels.append(labs)

    obj['text']=obj['text'].replace('\\n\\n',' ')
    obj['text']=obj['text'].replace('\\n',' ')
    
    obj['text']=obj['text'].replace('\n','')

    
    obj['text']=obj['text'].replace("\\","")
    
    obj['text']=obj['text'].replace("\'","")
    obj['text']=obj['text'].replace('"text":',"")
    obj['text']=obj['text'].replace('"','')
    obj['text']=obj['text'].strip()

  all_labels = list(set(list(itertools.chain(*all_labels))))
  print(len(all_labels))
  print("all labels: ", all_labels)
  return arr,all_labels
```

```python
def make_features_labels(arr,all_labels):
  "makes features and labels from the data"

  features=[]
  labels=[]

  for val in arr:
    features.append(val['text'])
    labs=list(np.zeros(len(all_labels)))
    for lab in val['labels']:
      labs[all_labels.index(lab)]=1
    labels.append(np.array(labs))
  labels=np.array(labels)
  return features,labels
```

```python
def extract_features_labels(path):
  "returns features, labels and all labels from the given data path"
  arr=read_text_file(path)
  arr,all_labels=clean_data(arr)

  features,labels=make_features_labels(arr,all_labels)
  return features,labels,all_labels
```
